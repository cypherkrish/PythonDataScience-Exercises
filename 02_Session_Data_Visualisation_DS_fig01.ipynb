```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import warnings
import datetime as dt
warnings.filterwarnings('ignore')
```

```python
dfm = pd.read_csv('https://raw.githubusercontent.com/cypherkrish/PythonDataScience/main/Data/matchs.csv')
dfd = pd.read_csv('https://raw.githubusercontent.com/cypherkrish/PythonDataScience/main/Data/deliveris.csv')
```

```python
fig1,ax1=plt.subplots(figsize=(15,8))
###Start code here
value_counts_df = dfm['season'].value_counts().reset_index()
value_counts_df.columns = ['season', 'matches']
sns.barplot(data = value_counts_df , x = 'season', y='matches', palette='husl')
ax1.set_title('Total Number of Matches hosted seasons wise') # fontsize=16)
ax1.set_xlabel('seasons') #, fontsize=14)
ax1.set_ylabel('matches') #, fontsize=14)
# Display the plot
plt.show()
```
