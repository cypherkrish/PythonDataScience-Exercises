```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import warnings
import datetime as dt
warnings.filterwarnings('ignore')
```

```python
dfm = pd.read_csv('https://raw.githubusercontent.com/cypherkrish/PythonDataScience/main/Data/matchs.csv')
dfd = pd.read_csv('https://raw.githubusercontent.com/cypherkrish/PythonDataScience/main/Data/deliveris.csv')
```

Draw a bar plot on matches won by team batting first.
Hint - Here batting first means team who won the toss and chose to bat.
- use seaborn.barplot()


```python
fig3,ax3=plt.subplots(figsize=(15,8))
###Start code here
value_counts_fig3 = dfm[(dfm['toss_winner'] == dfm['winner']) &  (dfm['toss_decision'] == 'bat')]['winner'].value_counts().reset_index()
value_counts_fig3.columns = ['Teams', 'WINS']
sns.barplot(data = value_counts_fig3 , x = 'Teams', y='WINS', palette='tab10')
ax3.set_title('Matches won by team batting first') # fontsize=16)
ax3.set_xlabel('Teams') #, fontsize=14)
ax3.set_ylabel('WINS') #, fontsize=14)
plt.xticks(rotation=90)
# Display the plot
plt.show()
```
