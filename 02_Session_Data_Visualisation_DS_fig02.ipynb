```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import warnings
import datetime as dt
warnings.filterwarnings('ignore')
```

```python
dfm = pd.read_csv('https://raw.githubusercontent.com/cypherkrish/PythonDataScience/main/Data/matchs.csv')
dfd = pd.read_csv('https://raw.githubusercontent.com/cypherkrish/PythonDataScience/main/Data/deliveris.csv')
```

Draw a count plot on the total number of games hosted in each stadium.

use seaborn.countplot()

```python
dfm = dfm.replace(to_replace="M Chinnaswamy Stadium",  value="M. Chinnaswamy Stadium")
```

```python
fig2,ax2=plt.subplots(figsize=(15,8))

sns.countplot(x = 'venue', data = dfm , order=dfm['venue'].value_counts(ascending=False).index, palette='Set1')
ax2.set_title('No. of games hosted in each stadium') # fontsize=16)
ax2.set_xlabel('venue') #, fontsize=14)
ax2.set_ylabel('count') #, fontsize=14)
plt.xticks(rotation=90)
plt.show()
```
